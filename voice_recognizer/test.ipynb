```python
import torch.jit
import os
import torchaudio
model = torch.jit.load("voice_recognizer.pt")
model.eval()
```

```python
test_path = './SpeechCommands/speech_commands_v0.02/stop/0b56bcfe_nohash_1.wav'
```

```python
def load_waveform(path: str, exp_sample_rate: int):
    waveform, sample_rate = torchaudio.load(path)
    if exp_sample_rate != sample_rate:
        raise ValueError(f"sample rate should be {exp_sample_rate}, but got {sample_rate}")
    return waveform
```

```python
waveform = load_waveform(test_path, 16_000)
```
